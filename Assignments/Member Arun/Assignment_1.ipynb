```python
import seaborn as sns
import numpy as np
import pandas
import matplotlib.pyplot as plt
```

```python
# Reading the .csv file
dataset = pandas.read_csv("House Price India.csv")
```

```python
dataset.info()
```

```python
# 1.Univariant
sns.countplot(data=dataset,y="number of bedrooms")
```

```python
# 2.Bivariant
sns.scatterplot(data=dataset,x="number of bathrooms", y='number of bedrooms')
```

```python
# 3.multivariant
sns.heatmap(dataset.corr())
```
